```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import pandas as pd
```

```python
from datetime import datetime

def get_ts():
    return datetime.now().strftime("%Y-%m-%d %H:%M:%S")
```

```python
df_treinamento = pd.read_csv("../bases/ciha_treinamento_sample10.csv") #.sample(frac=0.01, random_state=42)
```

```python
df_treinamento.count()
```

```python
df_treinamento.head()
```

```python
df_treinamento.dtypes
```

```python
categoricas_int = [
    "sexo",
    "dt_atend_dow",
    "feriado",
    "diag_princ_capitulo_numero",
    "proc_rea",
    "dt_atend_mes",
    "dt_nasc_mes",
    "dt_saida_mes"
]

categoricas_str = [
    "diag_princ",
    "munic_res_sigla_uf",
    "munic_mov_sigla_uf",
    "munic_res_nome_uf",
    "munic_mov_nome_uf",
    "diag_princ_categoria"
]

numericas = [
    "dt_atend_dia",
    "dt_atend_ano",
    "dt_atend_woy",
    "dt_nasc_dia",
    "dt_nasc_ano",
    "dt_nasc_woy",
    "dt_saida_dia",
    "dt_saida_ano",
    "dt_saida_woy",
    "uti_int_to",
    "qt_proc",
    "idade",
    "dias_perm",
    "feriados_qtd",
    "feriados_prox_qtd",
    "feriados_prox_dist",
    "munic_res_populacao_residente",
    "munic_mov_populacao_residente",
    "munic_res_area_unidade_territorial",
    "munic_mov_area_unidade_territorial",
    "munic_res_idhm",
    "munic_mov_idhm"
]

target = "morte"
y = df_treinamento[target]
X = df_treinamento.drop(columns=[target])
```

```python
print(get_ts(), "Inputs...")
inputs = {}
for col in numericas + categoricas_int + categoricas_str:
    print(get_ts(), col)
    inputs[col] = tf.keras.Input(shape=(1,), name=col, dtype=tf.float32 if col in numericas or col in categoricas_int else tf.string)
```

```python
print(get_ts(), "Numerics...")
num_features = []
for col in numericas:
    print(get_ts(), col)
    norm = layers.Normalization()
    norm.adapt(df_treinamento[[col]])
    num_features.append(norm(inputs[col]))
```

```python
print(get_ts(), "Categorics STR...")
cat_features = []
for col in categoricas_str:
    print(get_ts(), col)
    lookup = layers.StringLookup(output_mode="one_hot")
    lookup.adapt(df_treinamento[col])
    cat_features.append(lookup(inputs[col]))
```

```python
print(get_ts(), "Categorics INT...")
for col in categoricas_int:
    print(get_ts(), col)
    lookup2 = layers.IntegerLookup(output_mode="one_hot")
    lookup2.adapt(df_treinamento[col])
    cat_features.append(lookup2(inputs[col]))
```

```python
len(numericas + categoricas_str + categoricas_int)
```

```python
x = layers.concatenate(num_features + cat_features)
x = layers.Dense(128, activation='relu')(x)
x = layers.Dropout(0.3)(x)
x = layers.Dense(64, activation='relu')(x)
x = layers.Dropout(0.3)(x)
x = layers.Dense(32, activation='relu')(x)
x = layers.Dropout(0.3)(x)
output = layers.Dense(1, activation='sigmoid')(x)

# Modelo final
model = keras.Model(inputs=inputs, outputs=output)

# Compilar
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy', tf.keras.metrics.AUC(name="auc")])
```

```python
model.save("models/model1.keras")
model.save_weights("models/modelo1_weights.tf", save_format='tf')
```

```python
df_teste = pd.read_csv("../bases/ciha_teste_sample10.csv") #.sample(frac=0.01, random_state=42)
```

```python
print(get_ts(), "Inputs...")
inputs_teste = {}
for col in numericas + categoricas_int + categoricas_str:
    print(get_ts(), col)
    inputs_teste[col] = tf.keras.Input(shape=(1,), name=col, dtype=tf.float32 if col in numericas or col in categoricas_int else tf.string)

print(get_ts(), "Numerics...")
num_features_teste = []
for col in numericas:
    print(get_ts(), col)
    norm = layers.Normalization()
    norm.adapt(df_teste[[col]])
    num_features_teste.append(norm(inputs[col]))

print(get_ts(), "Categorics STR...")
cat_features_teste = []
for col in categoricas_str:
    print(get_ts(), col)
    lookup = layers.StringLookup(output_mode="one_hot")
    lookup.adapt(df_teste[col])
    cat_features_teste.append(lookup(inputs[col]))

print(get_ts(), "Categorics INT...")
for col in categoricas_int:
    print(get_ts(), col)
    lookup2 = layers.IntegerLookup(output_mode="one_hot")
    lookup2.adapt(df_teste[col])
    cat_features_teste.append(lookup2(inputs[col]))

x_teste = layers.concatenate(num_features_teste + cat_features_teste)
```

```python
x_teste_dict = {
    col: df_teste[col].values for col in numericas + categoricas_int + categoricas_str
}
```

```python
predicoes = model.predict(x_teste_dict)
```

```python
if len(predicoes.shape) == 2 and predicoes.shape[1] == 1:
    df_teste["predicao"] = predicoes[:, 0]
else:
    df_teste["predicao"] = predicoes
```

```python
df_teste["classe_predita"] = (df_teste["predicao"] >= 0.33).astype(int)
```

```python
df_teste.head()
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score

acc = accuracy_score(df_teste["morte"], df_teste["classe_predita"])
prec = precision_score(df_teste["morte"], df_teste["classe_predita"])
rec = recall_score(df_teste["morte"], df_teste["classe_predita"])
f1 = f1_score(df_teste["morte"], df_teste["classe_predita"])
auc = roc_auc_score(df_teste["morte"], df_teste["predicao"])

# Mostrar os resultados
print(f"Acurácia:  {acc:.4f}")
print(f"Precisão:  {prec:.4f}")
print(f"Recall:    {rec:.4f}")
print(f"F1-score:  {f1:.4f}")
print(f"AUC:       {auc:.4f}")
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

cm = confusion_matrix(df_teste["morte"], df_teste["classe_predita"])

# Plote
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=[0, 1])
disp.plot(cmap=plt.cm.Blues,  values_format='d')
plt.title("Matriz de Confusão")
plt.grid(False)
plt.show()
```

```python
from sklearn.metrics import roc_curve, auc

# Calcular pontos da curva ROC
fpr, tpr, thresholds = roc_curve(df_teste["morte"], df_teste["predicao"])
roc_auc = auc(fpr, tpr)

# Plote
plt.figure()
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f"ROC curve (AUC = {roc_auc:.4f})")
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--', label="Aleatório")
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel("Taxa de Falsos Positivos")
plt.ylabel("Taxa de Verdadeiros Positivos")
plt.title("Curva ROC")
plt.legend(loc="lower right")
plt.grid(True)
plt.show()
```
