```python
import builtins as b
from datetime import datetime, timedelta, date
import sys, traceback
from delta.tables import DeltaTable

from scipy.cluster.hierarchy import linkage, dendrogram
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from pyspark.ml.clustering import KMeans
from pyspark.ml.feature import StringIndexer, VectorAssembler, OneHotEncoder
from pyspark.ml.evaluation import ClusteringEvaluator
from pyspark.ml import Pipeline
```

```python
df = spark.table("stage.ciha_atendimentos")

df.printSchema()
```

```python
df
```

```python
# date_format(col("dt_atend"), "yyyy-MM")
# date_trunc("month", col("dt_atend"))

df_mensal = df.filter(col("dt_atend").between(lit('2011-01-01').cast('date'), lit('2023-12-31').cast('date'))).withColumn("mes", date_trunc("month", col("dt_atend"))) \
                     .groupBy("mes").agg(
                         sum("total_atendimentos").alias("qtd_mes")
                     ).orderBy(col("mes").asc())
```

```python
df_mensal
```

```python
df_mensal_pd = df_mensal.toPandas()
```

```python
df_mensal_pd.head()
```

```python
df_mensal_pd.set_index('mes', inplace=True)
df_mensal_pd.index = pd.DatetimeIndex(df_mensal_pd.index, freq='MS')

df_mensal_pd.head()
```

```python
df_prophet = df_mensal_pd.copy()

df_prophet = df_prophet.rename(columns={"mes": "ds", "qtd_mes": "y"})

df_prophet["ds"] = pd.to_datetime(df_prophet["ds"])

df_prophet.head()
```

```python
from prophet import Prophet

model = Prophet()
model.fit(df_prophet)
```

```python
future = model.make_future_dataframe(periods=12, freq='MS')

forecast = model.predict(future)
```

```python
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail(12)
```

```python
model.plot(forecast)
plt.title("Previsão de Atendimentos Médicos")
plt.ticklabel_format(style='plain', axis='y')
plt.show()
```

```python
# Cria o dataframe com os dados de 2024 para comparar com o modelo
df_mensal_2024 = df.filter(col("dt_atend").between(lit('2024-01-01').cast('date'), lit('2024-12-31').cast('date'))).withColumn("mes", date_trunc("month", col("dt_atend"))) \
                     .groupBy("mes").agg(
                         sum("total_atendimentos").alias("qtd_mes")
                     ).orderBy(col("mes").asc())

df_mensal_2024_pd = df_mensal_2024.toPandas()

df_mensal_2024_pd.set_index('mes', inplace=True)
df_mensal_2024_pd.index = pd.DatetimeIndex(df_mensal_2024_pd.index, freq='MS')
```

```python
# Pega só os meses de 2024
previsoes_2024 = forecast[forecast['ds'].dt.year == 2024]
yhat_2024 = previsoes_2024.set_index('ds')['yhat']

# Alinhe com a série real
real_2024 = df_mensal_2024_pd.copy()
real_2024.index = pd.to_datetime(real_2024.index)
common_idx = real_2024.index.intersection(yhat_2024.index)

# Cálculo de métricas
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

mae = mean_absolute_error(real_2024.loc[common_idx], yhat_2024.loc[common_idx])
rmse = mean_squared_error(real_2024.loc[common_idx], yhat_2024.loc[common_idx], squared=False)
mape = (b.abs((real_2024.loc[common_idx] - yhat_2024.loc[common_idx].to_frame().rename(columns={"yhat": "qtd_mes"})) / yhat_2024.loc[common_idx].to_frame().rename(columns={"yhat": "qtd_mes"}))).mean() * 100
r2 = r2_score(real_2024.loc[common_idx], yhat_2024.loc[common_idx])

print(f"MAE: {mae:.2f}")
print(f"RMSE: {rmse:.2f}")
print(f"MAPE: {mape.iloc[0]:.2f}%")
print(f"R²: {r2:.4f}")
```

```python
previsoes_2024 = forecast[forecast['ds'].dt.year == 2024]
yhat_2024 = previsoes_2024.set_index('ds')['yhat']

real_2024.index = pd.DatetimeIndex(real_2024.index)
yhat_2024.index = pd.DatetimeIndex(yhat_2024.index)

common_index = real_2024.index.intersection(yhat_2024.index)
real_aligned = real_2024.loc[common_index]
previsto_aligned = yhat_2024.loc[common_index]
```

```python
import matplotlib.pyplot as plt

yhat_lower = previsoes_2024.set_index('ds')['yhat_lower']
yhat_upper = previsoes_2024.set_index('ds')['yhat_upper']

plt.figure(figsize=(12,6))
plt.plot(real_aligned, label="Real 2024", marker='o')
plt.plot(previsto_aligned, label="Previsto (Prophet)", marker='x')
plt.fill_between(yhat_2024.index, yhat_lower, yhat_upper, color='gray', alpha=0.2, label='Intervalo de Confiança')
plt.title("Comparação Real vs Previsto - Ano 2024")
plt.xlabel("Mês")
plt.ylabel("Atendimentos")
plt.xticks(rotation=45)
plt.grid()
plt.legend()
plt.tight_layout()

plt.ticklabel_format(style='plain', axis='y')

plt.show()
```
