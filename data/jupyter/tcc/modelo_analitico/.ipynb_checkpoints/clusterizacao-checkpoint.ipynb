```python
import builtins as b
from datetime import datetime, timedelta, date
import sys, traceback
from delta.tables import DeltaTable

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from pyspark.ml.clustering import KMeans
from pyspark.ml.feature import StringIndexer, VectorAssembler
from pyspark.ml.evaluation import ClusteringEvaluator
from pyspark.ml import Pipeline
```

```python
df = spark.table("stage.ciha_step3")

df.printSchema()
```

```python
df
```

```python
df.count()
```

```python
df_t = df.select(
    col("sexo").cast("integer").alias("sexo"), 
    col("morte").cast("integer").alias("morte"), 
    col("cod_idade").cast("integer").alias("cod_idade"),
    col("munic_mov_idhm").cast("double").alias("munic_mov_idhm"), 
    col("munic_mov_sigla_uf"), 
    col("diag_princ_categoria"), 
    col("proc_no_grupo")
).limit(1000)

df_t.printSchema()
```

```python
df_t
```

```python
schema = df_t.dtypes
colunas_string = [name for name, dtype in schema if dtype == "string"]
colunas_numericas = [name for name, dtype in schema if dtype in ("int", "double", "float")]

print(colunas_string)
print(colunas_numericas)
```

```python
indexadores = [
    StringIndexer(inputCol=col, outputCol=f"{col}_index", handleInvalid="keep")
    for col in colunas_string
]

colunas_features = colunas_numericas + [f"{col}_index" for col in colunas_string]

assembler = VectorAssembler(inputCols=colunas_features, outputCol="features")

pipeline = Pipeline(stages=indexadores + [assembler])
df_final = pipeline.fit(df_t).transform(df_t)

df_final.select("features").show(5, truncate=False)
```

```python
kmeans = KMeans(featuresCol='features', k=3)
modelo = kmeans.fit(df_final)
resultado = modelo.transform(df_final)

resultado.select("prediction").show()
```

```python
resultado.printSchema()
```

```python
cols_para_plot = colunas_features + ["prediction"]

resultado_pd = resultado.select(*cols_para_plot).toPandas()

plt.figure(figsize=(8, 6))
scatter = plt.scatter(
    resultado_pd["cod_idade"],
    resultado_pd["sexo"],
    c=resultado_pd["prediction"],
    cmap="viridis",
    s=50
)

plt.xlabel("Idade")
plt.ylabel("Sexo (indexado)")
plt.title("Clusters com KMeans (idade x sexo)")
plt.colorbar(scatter, label="Cluster")
plt.grid(True)
plt.show()
```
